```python
#Save all leaf nodes of a Binary tree in a Doubly Linked List by using Right node as Next node and Left Node as Previous Node.
# source: https://stackoverflow.com/questions/2598437/how-to-implement-a-binary-tree?utm_medium=organic&utm_source=google_rich_qa&utm_campaign=google_rich_qa
class Node:
    def __init__(self, val):
        self.l = None
        self.r = None
        self.v = val

class Tree:
    def __init__(self):
        self.root = None

    def getRoot(self):
        return self.root

    def add(self, val):
        if(self.root == None):
            self.root = Node(val)
        else:
            self._add(val, self.root)

    def _add(self, val, node):
        if(val < node.v):
            if(node.l != None):
                self._add(val, node.l)
            else:
                node.l = Node(val)
        else:
            if(node.r != None):
                self._add(val, node.r)
            else:
                node.r = Node(val)

    def find(self, val):
        if(self.root != None):
            return self._find(val, self.root)
        else:
            return None

    def _find(self, val, node):
        if(val == node.v):
            return node
        elif(val < node.v and node.l != None):
            self._find(val, node.l)
        elif(val > node.v and node.r != None):
            self._find(val, node.r)

    def deleteTree(self):
        # garbage collector will do this for us. 
        self.root = None

    def printTree(self):
        if(self.root != None):
            self._printTree(self.root)

    def _printTree(self, node):
        if(node != None):
            self._printTree(node.l)
            print (str(node.v) + ' ')
            self._printTree(node.r)

# source: http://ls.pwd.io/2014/08/singly-and-doubly-linked-lists-in-python/
class ListNode(object):
 
    def __init__(self, data, prev, next):
        self.data = data
        self.prev = prev
        self.next = next
 
 
class DoubleList(object):
 
    head = None
    tail = None
 
    def append(self, data):
        new_node = ListNode(data, None, None)
        if self.head is None:
            self.head = self.tail = new_node
        else:
            new_node.prev = self.tail
            new_node.next = None
            self.tail.next = new_node
            self.tail = new_node
            
 
    def remove(self, node_value):
        current_node = self.head
 
        while current_node is not None:
            if current_node.data == node_value:
                # if it's not the first element
                if current_node.prev is not None:
                    current_node.prev.next = current_node.next
                    current_node.next.prev = current_node.prev
                else:
                    # otherwise we have no prev (it's None), head is the next one, and prev becomes None
                    self.head = current_node.next
                    current_node.next.prev = None
 
            current_node = current_node.next
 
    def show(self):
        print ("Show list data:")
        current_node = self.head
        while current_node is not None:
            #print (current_node.prev.data if hasattr(current_node.prev, "data") else None)
            print (current_node.data)
            #print (current_node.next.data if hasattr(current_node.next, "data") else None)
 
            current_node = current_node.next
        print ("*"*50)

#Save all leaf nodes of a Binary tree in a Doubly Linked List by using Right node as Next node and Left Node as Previous Node.
# recurse through tree, insert current node into list, insert left child into list left, insert right child into list right, repeat for left and right children
def leavesToLinkedList(tree):
    result = DoubleList()
    leavesToLinkedList_recursive(tree.getRoot(), result)
    return result
def leavesToLinkedList_recursive(node, linkedList):
    if node is None:
        return None
    
    if node.l is None and node.r is None:
        linkedList.append(node.v)
    else:
        if node.l is not None:
            leavesToLinkedList_recursive(node.l, linkedList)
        if node.r is not None:
            leavesToLinkedList_recursive(node.r, linkedList)
            
binaryTree = Tree()
values = [5,3,7,2,9,0,3]
for value in values:
    binaryTree.add(value)
leaves = leavesToLinkedList(binaryTree)
leaves.show()
```

```python
#Given an array,find the maximum j – i such that arr[j] > arr[i]

def maxGap(array):
    #naive
    maxGapSize = 0
    lMin = [array[0]] * len(array)
    rMax = [array[-1]] * len(array)
    minElement = array[0]
    for index in range(1,len(array)):
        if array[index] < minElement:
            minElement = array[index]
        lMin[index] = minElement
    maxElement = array[-1]
    for index in range(len(array) - 2, -1, -1):
        if array[index] > maxElement:
            maxElement = array[index]
        rMax[index] = maxElement
        
    print(lMin, rMax)
    maxGapSize = 0
    i = 0
    j = 0
    while i < len(array) and j < len(array):
        if lMin[i] >= rMax[j]:
            i += 1
        else:
            diff = j - i
            if diff > maxGapSize:
                maxGapSize = diff
            j += 1
    return maxGapSize

array = [2,1,5,2,3,7,10,2,4,2,0]
result = maxGap(array)
print(result)
```
